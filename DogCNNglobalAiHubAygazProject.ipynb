```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Veri setinin bulunduğu klasör
data_folder = "/content/Images/"

# Veri setinin bir DataFrame'e yüklenmesi (örnek olarak, dosya adı ve etiket içeriyor varsayalım)
# Bu kısmı veri setinize göre düzenlemeniz gerekebilir
data = pd.DataFrame(columns=["Image_Name", "Label"])

for label in os.listdir(data_folder):
    label_folder = os.path.join(data_folder, label)
    if os.path.isdir(label_folder):
        for image_name in os.listdir(label_folder):
            data = data.append({"Image_Name": os.path.join(label, image_name), "Label": label}, ignore_index=True)

# Kategorik hedefi Label Encoding kullanarak sayısallaştırın
label_encoder = LabelEncoder()
data["Encoded_Label"] = label_encoder.fit_transform(data["Label"])

# Veri setini eğitim, test ve doğrulama setlerine bölün
X_train_names, X_temp_names, y_train, y_temp = train_test_split(data["Image_Name"], data["Encoded_Label"], test_size=0.3, random_state=42)
X_val_names, X_test_names, y_val, y_test = train_test_split(X_temp_names, y_temp, test_size=0.5, random_state=42)

# Resim verilerini yükleyerek normalleştirme işlemlerini yapın
image_size = (64, 64)  # Resim boyutlarına uygun olarak güncelleyin
batch_size = 32

train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_dataframe(
    dataframe=pd.DataFrame({'Image_Name': X_train_names, 'Encoded_Label': y_train}),
    directory=data_folder,
    x_col='Image_Name',
    y_col='Encoded_Label',
    target_size=image_size,
    batch_size=batch_size,
    class_mode='raw'
)

val_generator = val_datagen.flow_from_dataframe(
    dataframe=pd.DataFrame({'Image_Name': X_val_names, 'Encoded_Label': y_val}),
    directory=data_folder,
    x_col='Image_Name',
    y_col='Encoded_Label',
    target_size=image_size,
    batch_size=batch_size,
    class_mode='raw'
)

test_generator = test_datagen.flow_from_dataframe(
    dataframe=pd.DataFrame({'Image_Name': X_test_names, 'Encoded_Label': y_test}),
    directory=data_folder,
    x_col='Image_Name',
    y_col='Encoded_Label',
    target_size=image_size,
    batch_size=batch_size,
    class_mode='raw'
)
```
