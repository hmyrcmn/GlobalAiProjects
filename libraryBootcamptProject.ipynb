```python
class Library:
    def __init__(self):
        self.file = open("books.txt", "a+")

    def __del__(self):
        self.file.close()

    def list_books(self):
        self.file.seek(0)
        lines = self.file.read().splitlines()

        if not lines:
            print("Kütüphane boş.")
        else:
            print("Kütüphane İçeriği:")
            for line in lines:
                book_info = line.split(',')
                book_title = book_info[0]
                book_author = book_info[1]
                print(f"Kitap Başlığı: {book_title},\n  Yazar: {book_author}")

    def add_book(self):
        book_title = input("Enter the book title: ")
        book_author = input("Enter the author: ")
        release_date = input("Enter the release date: ")
        page_count = input("Enter the page count: ")

        book_info = f"{book_title},{book_author},{release_date},{page_count}\n"
        self.file.write(book_info)
        print(f"{book_title} by {book_author} has been added to the library.")

    def remove_book(self):
        book_title_to_remove = input("Enter the book title to remove: ")

        # Dosya içeriğini oku ve kitapları bir listeye ekle
        self.file.seek(0)
        lines = self.file.read().splitlines()

        if not lines:
            print("Kütüphane boş. Kaldırılacak kitap yok.")
        else:
            # Silinecek kitabın indeksini bul
            index_to_remove = None
            for index, line in enumerate(lines):
                book_info = line.split(',')
                book_title = book_info[0]
                if book_title == book_title_to_remove:
                    index_to_remove = index
                    break

            if index_to_remove is not None:
                # Kitabı listeden kaldır
                removed_book = lines.pop(index_to_remove)
                print(f"{removed_book} has been removed from the library.")

                # "books.txt" dosyasının içeriğini sil
                self.file.seek(0)
                self.file.truncate()

                # Liste elemanlarını "books.txt" dosyasına ekle
                for line in lines:
                    self.file.write(line + '\n')
            else:
                print(f"{book_title_to_remove} not found in the library.")

# Library sınıfından bir nesne oluştur
lib = Library()

# Menü
while True:
    print("\n*** MENÜ ***")
    print("1) Kitapları Listele")
    print("2) Kitap Ekle")
    print("3) Kitap Sil")
    print("4) Çıkış")

    choice = input("Lütfen bir seçenek girin (1-4): ")

    if choice == "1":
        lib.list_books()
    elif choice == "2":
        lib.add_book()
    elif choice == "3":
        lib.remove_book()
    elif choice == "4":
        print("Çıkış yapılıyor. İyi günler!")
        break
    else:
        print("Geçersiz giriş. Lütfen 1-4 arasında bir seçenek girin.")
```

# tarih giriş formatı düzenlendi


```python
from datetime import datetime

class Library:
    def __init__(self):
        self.file = open("books.txt", "a+")

    def __del__(self):
        self.file.close()

    def list_books(self):
        self.file.seek(0)
        lines = self.file.read().splitlines()

        if not lines:
            print("Kütüphane boş.")
        else:
            print("Kütüphane İçeriği:")
            for line in lines:
                book_info = line.split(',')
                book_title = book_info[0]
                book_author = book_info[1]
                release_date = book_info[2]
                page_count = book_info[3]
                print(f"Kitap Başlığı: {book_title}, Yazar: {book_author}, Yayın Tarihi: {release_date}, Sayfa Sayısı: {page_count}")

    def add_book(self):
        book_title = input("Kitap başlığını girin: ")
        book_author = input("Yazarı girin: ")

        # Tarih girişini doğrulama
        while True:
            release_date_input = input("Yayın tarihini girin (YYYY-MM-DD): ")
            try:
                release_date = datetime.strptime(release_date_input, "%Y-%m-%d").date()
                break
            except ValueError:
                print("Geçersiz tarih formatı. Lütfen YYYY-MM-DD formatında girin.")

        # Sayfa sayısını doğrulama
        while True:
            page_count_input = input("Sayfa sayısını girin: ")
            if page_count_input.isdigit():
                page_count = int(page_count_input)
                break
            else:
                print("Sayfa sayısı sadece rakamlardan oluşmalıdır.")

        book_info = f"{book_title},{book_author},{release_date},{page_count}\n"
        self.file.write(book_info)
        print(f"{book_title} - {book_author} tarafından eklenmiştir.")

    def remove_book(self):
        self.list_books()
        book_title_to_remove = input("Kaldırmak istediğiniz kitabın başlığını girin: ")

        # Dosya içeriğini oku ve kitapları bir listeye ekle
        self.file.seek(0)
        lines = self.file.read().splitlines()

        if not lines:
            print("Kütüphane boş. Kaldırılacak kitap yok.")
        else:
            # Silinecek kitabın indeksini bul
            index_to_remove = None
            for index, line in enumerate(lines):
                book_info = line.split(',')
                book_title = book_info[0]
                if book_title == book_title_to_remove:
                    index_to_remove = index
                    break

            if index_to_remove is not None:
                # Kitabı listeden kaldıra
                removed_book = lines.pop(index_to_remove)
                print(f"{removed_book} kütüphaneden kaldırıldı.")

                # "books.txt" dosyasının içeriğini sil
                self.file.seek(0)
                self.file.truncate()

                # Liste elemanlarını "books.txt" dosyasına ekle
                for line in lines:
                    self.file.write(line + '\n')
            else:
                print(f"{book_title_to_remove} kütüphanede bulunamadı.")

# Library sınıfından bir nesne oluştur
lib = Library()

# Menü
while True:
    print("\n*** MENÜ ***")
    print("1) Kitapları Listele")
    print("2) Kitap Ekle")
    print("3) Kitap Sil")
    print("4) Çıkış")

    choice = input("Lütfen bir seçenek girin (1-4): ")

    if choice == "1":
        lib.list_books()
    elif choice == "2":
        lib.add_book()
    elif choice == "3":
        lib.remove_book()
    elif choice == "4":
        print("Çıkış yapılıyor. İyi günler!")
        break
    else:
        print("Geçersiz giriş. Lütfen 1-4 arasında bir seçenek girin.")
```
